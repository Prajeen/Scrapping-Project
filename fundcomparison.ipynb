```python
import pandas as pd
from openpyxl import load_workbook
from datetime import date, datetime, timedelta
from openpyxl.utils.dataframe import dataframe_to_rows
```

```python
details_file_path = 'C:/Prajeen/My Docs/Equity/1.Equity Details/1.Fund Analysis/Fund Details/'
fundsExcel = 'elss_17_Feb_2018.xlsx'
```

```python
def getDfFromXl(excel):
    headers = list(excel.values)[0]
    content = list(excel.values)[1:]
    return pd.DataFrame(data=content,columns=headers)
```

```python
def getToday():
    return (date.today()-timedelta(1)).strftime('%d/%m/%Y')
```

```python
fundData = load_workbook(details_file_path+fundsExcel)
fundDf = getDfFromXl(fundData['funds'])
```

```python
returnPoints = ['1W Ret','1M Ret','3M Ret','6M Ret','1Y Ret','3Y Ret']
retursDFrame = fundDf[returnPoints]
dfMean = retursDFrame.mean()
dfRange = retursDFrame.max() - retursDFrame.min()
print(dfMean)
for val in returnPoints:
    newCol = val+'Norm'
    retursDFrame[newCol] = retursDFrame[val].apply(lambda x: (x-dfMean[val])/dfRange[val])

retursDFrame = retursDFrame.assign(normSum3YR = lambda x: x['1W RetNorm']*0.1 + x['1M RetNorm']*0.5 + x['3M RetNorm']*0.8 
                                                + x['6M RetNorm']*0.8 + x['1Y RetNorm']*0.9 + x['3Y RetNorm'])
retursDFrame = retursDFrame.assign(normSum1Y = lambda x: x['1W RetNorm']*0.2 + x['1M RetNorm']*0.6 + x['3M RetNorm']*0.8
                                                + x['6M RetNorm']*0.9+ x['1Y RetNorm'])
retursDFrame = retursDFrame.assign(normSum6M = lambda x: x['1W RetNorm']*0.25 + x['1M RetNorm']*0.7 + x['3M RetNorm']*0.9
                                                + x['6M RetNorm'])
```
